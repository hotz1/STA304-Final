```python
# Import statements (You my need to run "!pip install sportsipy" first)
!pip install sportsipy
!pip install sportsreference

import pandas as pd
import sportsipy
from sportsipy.nfl.teams import Teams

from google.colab import drive
```

```python
yeer = 2020

teams = Teams(yeer).dataframes
teams = teams.assign(season = yeer)
teams
```

```python
teams_df = pd.DataFrame()
for yr in range(1970, 2021):
  annual_teams = Teams(yr).dataframes
  annual_teams = annual_teams.assign(season = yr)
  teams_df = teams_df.append(annual_teams, ignore_index = True)
```
